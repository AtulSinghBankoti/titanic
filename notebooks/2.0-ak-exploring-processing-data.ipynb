```python
#functon to extract the title from the name
def GetTitle(name):
    first_name_with_title=name.split(',')[1]
    title=first_name_with_title.split('.')[0]
    title=title.strip().lower()
    return title
```

```python
# function to extract the title from the name
def GetTitle(name):
    title_group={'mr':'Mr', 
                 'mrs':'Mrs',
                 'miss':'Miss',
                 'master':'Master',
                 'don':'Sir', 
                 'rev':'Sir',
                 'dr':'Officer', 
                 'mme':'Mrs',
                 'ms':'Mrs',
                'major':'Officer', 
                 'lady':'Lady',
                 'sir':'Sir',
                 'mlle':'Miss',
                 'col':'Officer',
                 'capt':'Officer',
                 'the countess':'Lady',
                   'jonkheer':'Sir',
                 'dona':'Lady'}
    first_name_with_title=name.split(',')[1]
    title=first_name_with_title.split('.')[0]
    title=title.strip().lower()
    return title_group[title]
```

```python
#create Title feature
df['Title']=df.Name.map(lambda x:GetTitle(x))
```
